```python
# ÏûëÏóÖÏóê ÌïÑÏöîÌïú Ìå®ÌÇ§ÏßÄÎ•º Î∂àÎü¨ÏòµÎãàÎã§
from selenium import webdriver
from selenium.webdriver.chrome.service import Service as ChromeService
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By

url = 'https://www.diningcode.com/list.dc?query=%EC%9D%98%EC%A0%95%EB%B6%80%20%EB%A7%9B%EC%A7%91'
# Chrome Î∏åÎùºÏö∞Ï†ÄÎ•º Ïò§ÌîàÌï©ÎãàÎã§
driver = webdriver.Chrome()
driver.get(url)
```

```python
header_list = []
score_list = []
category_list = []
review_list = []

# ÌäπÏ†ï ÏöîÏÜåÏùò ÌÖçÏä§Ìä∏ Ï∂îÏ∂ú
header = driver.find_elements(By.CLASS_NAME, 'InfoHeader')
score = driver.find_elements(By.CLASS_NAME, 'Score')
category = driver.find_elements(By.CLASS_NAME, 'Category')
review = driver.find_elements(By.CLASS_NAME, 'Review')

# Í∞Å ÏöîÏÜåÏùò ÌÖçÏä§Ìä∏Î•º Ï∂úÎ†•
for element in header:
    header_list.append(element.text)

for element in score:
    score_list.append(element.text)

for element in category:
    category_list.append(element.text)

for element in review:
    review_list.append(element.text)

# Î∏åÎùºÏö∞Ï†Ä Ï¢ÖÎ£å
driver.quit()
```

```python
category_list = [item.replace('\n', '') for item in category_list]
category_list = [item.replace(' ', '') for item in category_list]
temp_list = []
for element in header_list:
    idx = element.find('.')
    temp_list.append(element[idx+2:])
header_list = temp_list
```

```python
string = ''

for i in range(100):
    string += f'ÏãùÎãπÏù¥Î¶Ñ:{header_list[i]}\n'
    string += f'ÌèâÏ†ê:{score_list[i]}\n'
    string += f'ÏùåÏãùÏ¢ÖÎ•ò:{category_list[i]}\n'
    string += f'Î¶¨Î∑∞:{review_list[i]}\n'
```

```python
from docx import Document

# Î≥ÄÌôòÌï† Î¨∏ÏûêÏó¥

# Word Î¨∏ÏÑú Í∞ùÏ≤¥ ÏÉùÏÑ±
doc = Document()

# Î¨∏ÏûêÏó¥ÏùÑ Î¨∏ÏÑúÏóê Ï∂îÍ∞Ä
doc.add_paragraph(string)

# Î¨∏ÏÑúÎ•º .docx ÌååÏùºÎ°ú Ï†ÄÏû•
doc.save('output.docx')

print("Word Î¨∏ÏÑúÍ∞Ä 'output.docx'Î°ú Ï†ÄÏû•ÎêòÏóàÏäµÎãàÎã§.")
```
