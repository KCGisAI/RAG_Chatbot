```python
# 작업에 필요한 패키지를 불러옵니다
from selenium import webdriver
from selenium.webdriver.chrome.service import Service as ChromeService
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By

url = 'https://www.diningcode.com/list.dc?query=%EC%9D%98%EC%A0%95%EB%B6%80%20%EB%A7%9B%EC%A7%91'
# Chrome 브라우저를 오픈합니다
driver = webdriver.Chrome()
driver.get(url)
```

```python
header_list = []
score_list = []
category_list = []
review_list = []

# 특정 요소의 텍스트 추출
header = driver.find_elements(By.CLASS_NAME, 'InfoHeader')
score = driver.find_elements(By.CLASS_NAME, 'Score')
category = driver.find_elements(By.CLASS_NAME, 'Category')
review = driver.find_elements(By.CLASS_NAME, 'Review')

# 각 요소의 텍스트를 출력
for element in header:
    header_list.append(element.text)

for element in score:
    score_list.append(element.text)

for element in category:
    category_list.append(element.text)

for element in review:
    review_list.append(element.text)

# 브라우저 종료
driver.quit()
```

```python
category_list = [item.replace('\n', '') for item in category_list]
category_list = [item.replace(' ', '') for item in category_list]
temp_list = []
for element in header_list:
    idx = element.find('.')
    temp_list.append(element[idx+2:])
header_list = temp_list
```

```python
string = ''

for i in range(100):
    string += f'식당이름:{header_list[i]}\n'
    string += f'평점:{score_list[i]}\n'
    string += f'음식종류:{category_list[i]}\n'
    string += f'리뷰:{review_list[i]}\n'
```

```python
from docx import Document

# 변환할 문자열

# Word 문서 객체 생성
doc = Document()

# 문자열을 문서에 추가
doc.add_paragraph(string)

# 문서를 .docx 파일로 저장
doc.save('output.docx')

print("Word 문서가 'output.docx'로 저장되었습니다.")
```
